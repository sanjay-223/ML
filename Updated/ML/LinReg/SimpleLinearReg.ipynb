```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

x = np.random.uniform(0, 10, 20)
slope_true, intercept_true = 3.5, 7.2
y = slope_true * x + intercept_true 


plt.scatter(x, y, label='Generated Data')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.legend()
plt.title('Generated Simple Linear Regression Dataset')
plt.show()
```

```python
m =  np.sum((x - np.mean(x))*(y-np.mean(y)))/(np.sum(x-np.mean(x))**2)
```
