```python
import pandas as pd

df = pd.read_csv('data.csv')
df = df.drop(df.columns[df.columns.str.contains('Unnamed')], axis=1)
op = dict(df['Play Golf'].value_counts())
rows = df.shape[0]
```

```python
def find_probs(out_label):
    prob = {}
    for column in df.columns[:-1]:
        column_d = {}
        for class_ in df[column].unique():
            t_d = {}
            for output in df[out_label].unique():
                occurrences = ((df[column] == class_) & (df[out_label] == output)).sum()
                t_d[output] = occurrences/op[output]
            prob[str(class_)] = t_d
    prob_df = pd.DataFrame(prob)
    return prob_df
```

```python
def predict(X,prob_df):
    p_yes = op['Yes']/rows
    p_no = op['No']/rows
    for attr in X:
        p_yes *= prob_df[attr]['Yes']
        p_no *= prob_df[attr]['No']
    P_yes = round(p_yes/(p_yes+p_no),2)
    P_no = round(p_no/(p_yes+p_no),2)
    return "Yes" if P_yes >= P_no else "No"
    
```

```python
prob_df = find_probs('Play Golf')
predict(['Rainy','Cool','Normal','False'],prob_df)
```

```python
def accuracy():
    count = 0
    for index,row in df.iterrows():
        r = list(map(str,row))
        if predict(r[:-1]) == r[-1]:
            count += 1
        
    print(count)
```
